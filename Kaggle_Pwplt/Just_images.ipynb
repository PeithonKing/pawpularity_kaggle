```python
import tensorflow as tf
import numpy as np
import pandas as pd
import PIL as pil
from keras_preprocessing.image import ImageDataGenerator
import tensorflow.keras.layers as tfl
```

```python
df=pd.read_csv("paths_pwp.csv")
df
```

```python
datagen=ImageDataGenerator()
train_generator=datagen.flow_from_dataframe(dataframe=df,
                                            directory="./train/",
                                            x_col="Id", y_col="Pawpularity", 
                                            class_mode="raw", 
                                            target_size=(150,150), 
                                            batch_size=32)
```

```python
model=tf.keras.Sequential([
    tfl.Conv2D(16,(3,3),input_shape=(150,150,3),activation='relu'),
    tfl.Conv2D(16,(3,3),activation='relu'),
    tfl.BatchNormalization(axis=3),
    tfl.MaxPool2D((2,2)),
    tfl.Conv2D(32,(3,3),activation='relu'),
    tfl.Conv2D(32,(3,3),activation='relu'),
    tfl.BatchNormalization(axis=3),
    tfl.MaxPool2D((2,2)),
    tfl.Conv2D(64,(3,3),activation='relu'),
    tfl.Conv2D(64,(3,3),activation='relu'),
    tfl.BatchNormalization(axis=3),
    tfl.MaxPool2D((2,2)),
    tfl.Conv2D(64,(3,3),activation='relu'),
    tfl.Conv2D(64,(3,3),activation='relu'),
    tfl.BatchNormalization(axis=3),
    tfl.MaxPool2D((2,2)),
    tfl.Flatten(),
    tfl.Dense(1024,activation='relu'),
    tfl.Dense(1,activation='relu')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(
        learning_rate=0.01,
        beta_1=0.9,
        beta_2=0.999),
        loss='MeanSquaredError',
        metrics=['RootMeanSquaredError']
)
```

```python
model.fit(
        train_generator,
        steps_per_epoch=256,
        epochs=500
)
```
